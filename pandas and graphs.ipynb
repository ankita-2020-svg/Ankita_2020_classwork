```python
#Filter data(only wheat)
wheat_data=df[df["Crop"]=="Wheat"]
print("\nWheat data:")
print(wheat_data)
```

```python
# Group by Year and sum production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nYearly Total Production:")
print(yearly_production)
```

```python
#scatter plot- Rainfall vs Production
import matplotlib.pyplot as plt

plt.scatter(df["Rainfall_mm"],df["Production_Tons"])
plt.xlabel("Rainfall(mm)")
plt.ylabel("Production(tons)")
plt.title("Rainfall vs. Production")
plt.show()
```

```python
# Bar graph- Rainfall vs Production
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
plt.bar(df["Rainfall_mm"],df["Production_Tons"],color='purple')

plt.xlabel("Rainfall(mm)")
plt.ylabel("Production(tons)")
plt.title("Rainfall vs. Production")
plt.show()
```

```python
print(yearly_production)
print(yearly_production.index)
print(yearly_production.values)
```

```python
# Bar graph
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(yearly_production.index,yearly_production.values,color='green')
plt.xticks(yearly_production.index )
plt.xlabel("Year")
plt.ylabel("Total Production(tons)")
plt.title("Total Production per year")
plt.show()
```

```python
#Line plot- Production over years without for loop
#Filter data for each crop
wheat=df[df["Crop"]=="Wheat"]  #wheat is the dataframe and Wheat is the crop name taken from he table
rice=df[df["Crop"]=="Rice"]

plt.figure()

#plot separately
plt.plot(wheat["Year"],wheat["Production_Tons"],label="Wheat")
plt.plot(rice["Year"],rice["Production_Tons"],label="Rice")

plt.xlabel("Year")
plt.ylabel("Production(Tons)")
plt.title("Crop production over years")
plt.legend()
plt.show()
```

```python
#Bar plot- Production by crop(Single year eg:2022)
#Filter year 2022
data_2022=df[df["Year"]==2022]
print(data_2022)

plt.figure(figsize=(2,2))
plt.bar(data_2022["Crop"],data_2022["Production_Tons"],color=['skyblue','pink'])

plt.xlabel("Crop")
plt.ylabel("Production(tons)")
plt.title("Production by crop in the year 2022")
plt.show()
```

```python
#Custom style Bar plot
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
bars=plt.bar(yearly_production.index,
             yearly_production.values,
             color="green",
             edgecolor='black')
for bar in bars:
  height = bar.get_height()
  plt.text(bar.get_x()+bar.get_width()/2,
           height,
           str(height),
           ha='center',
           va='bottom')

  plt.xlabel("Year")
  plt.ylabel("Total Production(tons)")
  plt.title("Yearly Production with Values")
  plt.show()

```
