```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read dataset
data=pd.read_csv("/content/iris.csv")
data
```

```python
#dataframe["column_name"].unique()-gives unique values in the column
data["Species"].unique()
```

```python
#dataframe["column_name"].value_counts()- gives the count of each unique value in the column
data["Species"].value_counts()
```

```python
#Separate dependent and independent variables (input/output)
x=data.iloc[:,:-1].values
y=data.iloc[:,4].values
print(x)
print(y)
```

```python
#Split dataset into train and test method
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#Create decision tree classifier object

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=None,min_samples_split=5,min_samples_leaf=2)
```

```python
#Train model ('fit' method)
#model=obj.fit(x_train,y_train)
model.fit(x_train,y_train)
print(model)
```

```python
#Predict the testing model

y_pred= model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#Visualization

from sklearn import tree
fn=['sepal length(cm)','sepal width(cm)','petal length(cm)','petal width(cm)']
cn=['setosa','versicolor','virginica']
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4), dpi=300)
tree.plot_tree(model,  # model is an object
              feature_names= fn,  # column name created as a list is sent here
              class_names=cn,  #class means each labels
              filled= True);   #we want everything to be shown here
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
y_pred_nb = nb.predict(x_test)
cmnb = confusion_matrix(y_test, y_pred_nb)
print(cmnb)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Accuracy:", accuracy_nb)


```

```python
#Support Vector Machine (SVM)- regression, clasification

from sklearn.svm import SVC  # svc-classification, svr- regression

#Create SVM clasifier object
#model=SVC(kernel='rbf', C=1.0, gamma='scale', random_state=0)
#model= SVC(kernel='linear') #hyperparameter tuning
model= SVC(kernel='poly',degree=3)
#model= SVC(kernel='rbf)
#Training the model
model.fit(x_train, y_train)
print(model)
```

```python
#predicting the test set results
y_pred_svm= model.predict(x_test)
print(y_pred_svm)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm_svc= confusion_matrix(y_test,y_pred_svm)
print(cm_svc)
```

```python
#Accuracy
from sklearn.metrics import accuracy_score
print("SVM",accuracy_score(y_test,y_pred_svm))
```
