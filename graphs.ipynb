```python
#pie chart
import matplotlib.pyplot as plt
crops=['Wheat','Rice','Maize','Sorghum','Bajra']
yields=[90,180,200,120,150]
separated=(0,0,.1,0,0)   #the one for which .1 has been written will get separated from others by that much distance. We can put any decimal values
plt.pie(yields, labels=crops, autopct='%1.1f%%', explode=separated)
#plt.axis('equal')
plt.title("Yield percentage of crops")
plt.show()
```

```python
import matplotlib.pyplot as plt
portions=[10,5,4,3,2]
elements= ['Sky','Air','Earth','Water','Fire']
cols=['c','m','g','b','r']

plt.pie(portions,
        labels=elements,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(.1,0,0,0,0),
        autopct='%1.2f%%'  #displaying % values, the no. after decimal represents number of digits after decimal to be represented
        )

plt.title("Panchatatva")
plt.show()
```

```python
import numpy as np
years=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
open_sensex=[17473,20622,15535,19513,21222,27486,26102,26711,34060,36162,41349]
close_sensex=[20509,15455,19427,21171,27499,26118,26626,34057,36068,41254,33605]
ind=np.arange(len(years))
print(ind)
width=0.3
#fig= plt.figure(figsize=(6,6)) #1st 6= width and 2nd 6= height
plt.bar(ind,open_sensex,width,label="open")
plt.bar(ind+width,close_sensex,width,label="close",color='green')
plt.title("Sensex graph per year")
plt.ylabel("sensex")
plt.xlabel("year")
plt.xticks(ind, years, rotation=70)
plt.legend() # Create legend & Show graphic
plt.show() # Show
```
