```python
import pandas as pd

data={
    'Farm': ['Farm A','Farm B','Farm C','Farm D'],
    'Crop': ['Wheat','Rice','Wheat','Sorghum'],
    'Area_ha': [10,15,20,36],
    'Production_tons': [30,40,50,60]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Add a new column
df["Yield_tons_per_ha"]=df["Production_tons"]/df["Area_ha"]  #permanently changes the data farme
print(df)
```

```python
#slicing of columns
df.iloc[:][0:2]
```

```python
#filter data
#farms growing wheat
print(df['Crop']== 'Wheat')
print()
wheat_farms=df[df['Crop']=='Wheat']
print(wheat_farms)
```

```python
#Farms with yield >3
high_yield=df[df["Yield_tons_per_ha"]>=3]
print(high_yield)
```

```python
#Sort farms by by highest
sorted_df=df.sort_values(by="Production_tons",ascending=True)
print(sorted_df)
```

```python
#GroupBy operation- Avg yield per crop
avg_yield= df.groupby("Crop")["Yield_tons_per_ha"].mean()  #gives mean value of yield of wheat
print(avg_yield)
```

```python
avg_yield= df.groupby("Crop")["Yield_tons_per_ha"].sum() #gives sum of yields of wheat
print(avg_yield)
```
