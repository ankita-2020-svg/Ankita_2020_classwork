```python
#Math Library
import math as m
print("exp(-200)=",m.exp(-200)) #exponential function
print("log of 100 with base 2=",m.log(100,2))
print("log of 100 with base 10=",m.log(100,10))
print("cos value of 30=",m.cos(30))
print("sin value of 30=",m.sin(30))
print("tan value of 30=",m.tan(30))
print("square root of 324=",m.sqrt(324))
print("ceiling value of 89.9=",m.ceil(89.9))
print("floor value of 89.9=",m.floor(89.9))
```

```python
#Functions
#with parameter and without return type
def sumfn(a,b):
  add= a+b
  print(add)

m= 10
n=20
sumfn(m,n)
```

```python
#with parameter and with return type
def sum(a,b):
  add=a+b
  return add

m=10
n=20
print(sum(m,n))
```

```python
#without parameter and without return type
def sum():
  a=10
  b=20
  add=a+b
  print(add)
sum()
```

```python
#without parameter and with return type
def sum():
  a=10
  b=20
  add=a+b
  return add

print(sum())
```
