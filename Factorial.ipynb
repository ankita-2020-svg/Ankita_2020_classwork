```python
# finding max of the two numbers using max function
def max(n1,n2):
  if n1>n2:
    result=n1
  else:
    result=n2
  return result

x=10
y=20
z=max(x,y)
print(z)
```

```python
#finding factorial using function method
# with parameter and with return type
def fact(m):
  fact=1
  if(m>0):
    for i in range(1,m+1):
      fact=fact*i
    return fact
  else:
    print("Enter a valid no.")

n=int(input("Enter a number="))
a=fact(n)
print("Factorial of the number",a)
```
